```python
# Crawl Data

filename = 'crawling.csv'
search_keyword = 'program makan bergizi gratis lang:in until:2023-12-30 since:2023-01-01' #untuk keyword bisa diganti
# search_keyword = 'program makan bergizi gratis lang:in until:2024-12-30 since:2024-01-01' #untuk keyword bisa diganti
# search_keyword = 'program makan bergizi gratis lang:in until:2025-12-30 since:2025-01-01' #untuk keyword bisa diganti
# search_keyword = 'program makan bergizi gratis lang:in until:2026-04-30 since:2023-01-01' #untuk keyword bisa diganti
limit = 5000

!npx --yes tweet-harvest@latest -o "{filename}" -s "{search_keyword}" -l {limit} --token "b2a1a3f23b8eedc96126e8f39d5904d8ee8f7e71"
```

```python
import pandas as pd

file_path = f"tweets-data/{filename}"
df = pd.read_csv(file_path)
df = df[['full_text', 'id_str', 'created_at', 'username', 'lang', 'location', 'quote_count', 'reply_count', 'retweet_count', 'favorite_count', 'tweet_url', 'image_url']]
df['created_at'] = pd.to_datetime(df['created_at'])
df.sort_values(by='created_at', ascending=True, inplace=True)
```

```python
import re

def extract_mentions(text):
    mentions = re.findall(r'@(\w+)', text)
    return mentions

df['mentions'] = df['full_text'].apply(extract_mentions)
df['mention_count'] = df['mentions'].apply(len)

def choose_target(mentions):
    if mentions:
        return mentions[0]
    else:
        return None

df['target'] = df['mentions'].apply(choose_target)

df.to_csv('mbg2023.csv') # Ganti nama file yang akan disimpan
len(df)
```
