```python
import pandas as pd
import re
from difflib import SequenceMatcher

# =========================
# 1. Load data
# =========================
twitter = pd.read_csv("akuntwitter_data.csv", sep=';')
media = pd.read_csv("dewanpers_data.csv")
```

```python
print(twitter.columns)
```

```python
# =========================
# 2. Cleaning function
# =========================
def clean_text(text):
    if pd.isna(text):
        return ""
    text = str(text).lower()
    text = re.sub(r'[^a-z0-9 ]', ' ', text)  # hapus simbol
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# bersihkan kolom
twitter["clean_username"] = twitter["username"].apply(clean_text)
media["clean_media"] = media["Nama Media"].apply(clean_text)
```

```python

# =========================
# 3. Fungsi similarity
# =========================
def similarity(a, b):
    return SequenceMatcher(None, a, b).ratio()
```

```python

# =========================
# 4. Deteksi kategori
# =========================
def detect_category(username):
    for media_name in media["clean_media"]:
        # cek substring langsung
        if media_name in username or username in media_name:
            return "Media Resmi"

        # cek similarity (threshold bisa diubah)
        if similarity(username, media_name) > 0.7:
            return "Media Resmi"

    return "Individu/Komunitas"
```

```python

# =========================
# 5. Apply ke dataset
# =========================
twitter["kategori"] = twitter["clean_username"].apply(detect_category)
```

```python

# =========================
# 6. Simpan hasil
# =========================
twitter.to_csv("twitter_kategori.csv", index=False)

print("✅ Selesai! File: twitter_kategori.csv")
```
