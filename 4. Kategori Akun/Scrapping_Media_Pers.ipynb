```python
import requests
import pandas as pd
from time import sleep

# endpoint API
url = "https://dewanpers.or.id/portal-layanan-json-pendataan"

# parameter awal
limit = 10
start = 0

all_data = []

while True:
    params = {
        "limit": limit,
        "start": start,
        "search": "",
        "type": "",
        "province_id": "",
        "status": "",
        "orderBy": "",
        "sort": ""
    }

    res = requests.get(url, params=params)
    data = res.json()

    content = data.get("content", [])

    if not content:
        print("Selesai, tidak ada data lagi.")
        break

    print(f"Ambil data start={start}... ({len(content)} rows)")

    for item in content:
        all_data.append({
            "No": len(all_data) + 1,
            "Nama Media": item.get("media_name"),
            "Penanggung Jawab": item.get("media_person_in_charge_name"),
            "Alamat Media": item.get("media_address"),
            "Provinsi": item.get("media_province_id"),
            "Badan Hukum": item.get("media_legal_entity_name"),
            "Status": item.get("user_status_type")
        })

    # naik ke halaman berikutnya
    start += limit

    # biar aman (hindari rate limit)
    sleep(0.5)

# simpan ke DataFrame
df = pd.DataFrame(all_data)

# mapping ID provinsi ke nama
province_map = {
    '11': 'Aceh','12': 'Sumatera Utara','13': 'Sumatera Barat','14': 'Riau','15': 'Jambi',
    '16': 'Sumatera Selatan','17': 'Bengkulu','18': 'Lampung','19': 'Kepulauan Bangka Belitung',
    '21': 'Kepulauan Riau','31': 'DKI Jakarta','32': 'Jawa Barat','33': 'Jawa Tengah',
    '34': 'DI Yogyakarta','35': 'Jawa Timur','36': 'Banten','51': 'Bali',
    '52': 'Nusa Tenggara Barat','53': 'Nusa Tenggara Timur','61': 'Kalimantan Barat',
    '62': 'Kalimantan Tengah','63': 'Kalimantan Selatan','64': 'Kalimantan Timur',
    '65': 'Kalimantan Utara','71': 'Sulawesi Utara','72': 'Sulawesi Tengah',
    '73': 'Sulawesi Selatan','74': 'Sulawesi Tenggara','75': 'Gorontalo',
    '76': 'Sulawesi Barat','81': 'Maluku','82': 'Maluku Utara','91': 'Papua','92': 'Papua Barat'
}

df["Provinsi"] = df["Provinsi"].astype(str).map(province_map)

# simpan ke CSV
df.to_csv("dewanpers_data.csv", index=False)

print("✅ Selesai! Total data:", len(df))
```
